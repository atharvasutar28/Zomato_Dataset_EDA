```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\asus\Desktop\Python\data\zomato.csv")
df.head()
```

# Data Cleaning
## Deleting Redundant Columns

```python
df = df.drop(['url','address', 'phone', 'menu_item', 'reviews_list','dish_liked'], axis=1)
df
```

## Renaming Columns

```python
df = df.rename(columns = {'name':'restaurant_names','rate':'rating','rest_type':'restaurant_type'})
df
```

## Checking & Droping NULL Values 

```python
df.isnull().sum()
```

```python
def rate_clean(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
df['rating'] = df['rating'].apply(rate_clean)
df['rating'].unique()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.shape
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','').astype(int)
df['approx_cost(for two people)'].unique()
```

# Data Visualization

```python
#Restaurants delivering Online or not
plot1 = sns.countplot(x='online_order',data=df)
plot1.set(xlabel = 'Online Booking',ylabel = 'Restaurant Count',title = 'Restaurants Delivering Online or Not')
```

```python
#Restaurants allowing table booking or not
plot2 = sns.countplot(x='book_table',data=df)
plot2.set(xlabel = 'Table Booking',ylabel = 'Restaurant Count',title = 'Restaurants allowing Table Booking or Not')
```

```python
#Table booking Rate vs Rate
plot3 = sns.violinplot(x=df['book_table'], y=df['rating'])
plot3.set(xlabel = 'Table Booking',ylabel = 'Rating',
          title = 'Table Booking(Available) Rating vs Table Booking(Not-Available) Rating')
```

```python
#Best Location
criteria = df.loc[(df['rating'] > 4.7) & (df['votes'] > 16000)]
plot = sns.countplot(x='location',data=criteria,palette='rainbow')
plot.set(title='Best Location')
```

```python
#Relation between Location and Rating
plt.figure(figsize=(20,5))
temp = df.groupby('location')['rating'].mean().sort_values(ascending=True).head()
sns.barplot(x=temp.index,y=temp)
plt.xticks(rotation='vertical')
```

```python
plt.figure(figsize=(20,5))
sns.countplot(x= 'restaurant_type' ,data=df)
plt.xticks(rotation='vertical')
```

```python
#Restaurant Type
plt.figure(figsize=(20,13))
sns.barplot(x='rating',y='restaurant_type',data=df,palette='rainbow')
```

```python
#Gaussian Rest type and Rating
plt.figure(figsize=(20,5))
data = df.groupby('restaurant_type')['rating'].mean().sort_values(ascending=True)
sns.barplot(x=data.index,y=data)
plt.xticks(rotation='vertical')
```

```python
#Types of Services
plt.figure(figsize=(20,13))
sns.barplot(x='rating',y='restaurant_type',data=df,palette='rainbow')
```

```python
#Relation between Type and Rating
plt.figure(figsize=(20,5))
data = df.groupby('restaurant_type')['rating'].mean().sort_values(ascending=True)
sns.barplot(x=data.index,y=data)
plt.xticks(rotation='vertical')
```

```python
plt.figure(figsize=(20,5))
data = df.groupby('location')['approx_cost(for two people)'].mean().sort_values(ascending=False)
sns.barplot(x=data.index,y=data,palette='rocket')
plt.xticks(rotation='vertical')
```

```python
#No. of restaurants on a Location
plt.figure(figsize=(20,5))
data = df.groupby('location')['restaurant_names'].count().sort_values(ascending = False)
sns.barplot(x=data.index,y=data,palette='plasma')
plt.xticks(rotation='vertical')
```

```python
#Most famous restaurant chains in Bengaluru
data = df.loc[df['location'].isin(['South Bangalore','East Bangalore','Central Bangalore','West Bangalore','North Bangalore'])]
plt.figure(figsize=(20,5))
data1 = data.groupby('restaurant_names')['rating'].mean().sort_values(ascending=False)
sns.barplot(x=data1.index,y=data1)
plt.xticks(rotation='vertical')
```
